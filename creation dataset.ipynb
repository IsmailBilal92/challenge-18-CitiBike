# Duration Time
we need to calculate the duration time for each ride from both column start at and end at

```python
data_summer2023.info()
```

```python
data_summer2023['ended_at '] = data_summer2023['ended_at'].apply(pd.to_datetime)
```

```python
data_summer2023['end_lat'].dtypes
```

```python
data_summer2023 = data_summer2023.astype({'started_at': 'datetime64', 'ended_at': 'datetime64' })
```

```python
data_summer2023.dtypes
```

```python
Bikedata_summer2023 = data_summer2023.iloc[:,:-1]
```

```python
Bikedata_summer2023.dtypes
```

```python
from datetime import datetime
```

```python
Bikedata_summer2023['started_time'] = Bikedata_summer2023['started_at'].dt.time
```

```python
Bikedata_summer2023['ended_time'] = Bikedata_summer2023['ended_at'].dt.time
```

```python
Bikedata_summer2023.dtypes
```

```python
Bikedata_summer2023 = Bikedata_summer2023.astype({'started_at': 'object', 'ended_at': 'object' })
```

```python
Bikedata_summer2023.dtypes
```

```python
from datetime import datetime
```

```python
Bikedata_summer2023['ride_duration'] = Bikedata_summer2023.apply(lambda row: (datetime.combine(datetime.min, row['ended_time']) - datetime.combine(datetime.min, row['started_time'])), axis=1)
```

```python
Bikedata_summer2023 = Bikedata_summer2023.astype({'ride_duration': 'string'})
```

```python
Bikedata_summer2023['ride_duration'] = Bikedata_summer2023['ride_duration'].str.split(' ').str[-1]

```

```python
Bikedata_summer2023
```

```python
Bikedata_summer2023.info()
```

```python
Bikedata_summer2023.dropna(subset=['start_lat', 'start_lng', 'end_lat', 'end_lng'], inplace=True)
```

```python
Bikedata_summer2023["distance"] = ''
```

```python
import geopy.distance
from geopy.distance import geodesic

for index, row in Bikedata_summer2023.iterrows():
    coords_1 = (row['start_lat'], row['start_lng'])
    coords_2 = (row['end_lat'], row['end_lng'])
    distance_km = geopy.distance.geodesic(coords_1, coords_2).km
    Bikedata_summer2023.at[index, 'distance'] = distance_km
```

```python
# Define a custom function to convert the time duration to minutes
def time_to_minutes(time_str):
    time_parts = time_str.split(':')
    hours, minutes, seconds = map(int, time_parts)
    total_minutes = hours * 60 + minutes + seconds / 60
    return total_minutes

# Apply the custom function to the DataFrame column and store the result as a new column
Bikedata_summer2023['ride_duration_minutes'] = Bikedata_summer2023['ride_duration'].apply(time_to_minutes)
```
